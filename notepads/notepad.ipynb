```python
from collections import defaultdict
import chess
```

```python
all_moves = defaultdict(int)
position = 0

# all knight and queen moves (rooks, bishops, pawns, and kings are redundant movements)
for starting_square in chess.SQUARES:
    for piece_symbol in ['N','Q']:
        piece = chess.Piece.from_symbol(piece_symbol)
        board.clear()
        board.set_piece_at(square=starting_square, piece=piece)
        for move in board.legal_moves:
            assert not all_moves[move.uci()], 'UCI ALREADY FOUND.. YOU FUCKED UP'
            all_moves[move.uci()] = position
            position += 1

# all pawn promotions (one side)
board = chess.Board(fen='qqqqqqqq/8/8/8/8/8/8/QQQQQQQQ')
for file in range(8):
    for rank, piece_symbol in [(6, 'P')]:
        starting_square = chess.square(file_index=file, rank_index=rank)
        piece = chess.Piece.from_symbol(piece_symbol)
        board.set_piece_at(square=starting_square, piece=piece)
        
        inc = 1 if rank == 6 else -1
        blocking_square = chess.square(file_index=file, rank_index = rank + inc)
        board.remove_piece_at(square=blocking_square)
        
        for move in board.legal_moves:
            if not move.promotion:
                continue
            assert not all_moves[move.uci()], 'UCI ALREADY FOUND.. YOU FUCKED UP'
            all_moves[move.uci()] = position
            position += 1
        
        
        board.remove_piece_at(square=starting_square)
        blocking_queen = 'Q' if rank == 1 else 'q'
        board.set_piece_at(square=blocking_square, piece=chess.Piece.from_symbol(blocking_queen))
        
# castling (one side)
for uci in ['O-O','O-O-O']:
    assert not all_moves[uci], 'UCI ALREADY FOUND.. YOU FUCKED UP'
    all_moves[uci] = position
    position += 1

        
print(len(all_moves))
```
